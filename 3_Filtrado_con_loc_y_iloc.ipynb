# por qué usar loc e iloc?
Permiten filtrar datos de manera mas específica.

Entonces, ¿por qué loc e iloc ?
En los casos que queremos filtrar también por columna. 

En este caso ponemos:

```df.loc[filas, columnas]```

```df.iloc[filas, columnas]```

La diferencia entre estos, es que en loc, en las columnas utilizamos una lista de las que queremos, por su nombre, en iloc debemos hacerlo por índices.


__Loc filtra segun un label mientras que iloc lo hace mediante indices.__


# .loc | [Filtra según un label](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html)

````python
df_books.loc[:]
#muestra todos los datos del dataFrame
````

## Rango de filas tomando en cuenta el start y el end

````python
df_books.loc[0:4]
````

## Filtrando por filas y columnas | df.loc[0:4, ['Name', 'Author'] ] -> filas, columnas

````python
df_books.loc[0:4, ['Name', 'Author']]
#filtra los datos de la fila que va de 0 a 4 y de las columnas Name y Author
````

### Modificar valores de una columna específica del dataFrame

````python
df_books.loc[:, ['Reviews']] * -1
#multiplica por -1 todos los valores de la columna Reviews
````

### Filtrar datos que cumplan una condición determinada

````python
df_books.loc[:, ['Author']] == 'JJ Smith' 
#muestra la columna Author con True en los valores que cumplen la condicion y False para los que no la cumplen
````

# .iloc | [Filtra mediante índices.](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html)

````python
df_books.iloc[:] 
#muestra todos los datos del dataframe
````

## Filtrar datos según los índices de las filas y las columnas | df.iloc[0:4, 0:2] -> filas, columnas

````python
df_books.iloc[:4, 0:2] 
#muestra los datos de las filas que van de 0 a 3 y las columnas con indices 0 y 1
````

### Buscar un dato específico | df.iloc[\#fila, \#columna]

````python
df_books.iloc[1,3] 
#muestra el dato alojado en la fila 1 columna 3
````

# DIFERENCIA IMPORTANTE ENTRE LOC E ILOC

Más adelante veremos esto a detalle, pero lo que haremos es añadir filas a nuestro dataframe, pegando df_books bajo él mismo.

De lo anteriormente descrito quedamos con esto:

````python
df_books.append(df_books) # 1100 filas
````

LOC llama al indice o label, como pegamos el dataframe con su copia, aparece la fila consultada 2 veces:

````python
#llama al indice
df_books.append(df_books).loc[540]
````

como el indice de las filas van hasta el 549, "llamar la fila 550" es imposible aunque claramente ésta existe:

````python
#Salta Error
df_books.append(df_books).loc[550]
````

Con ILOC si podemos llamar las filas por su Número real y no por su índice

````python
#llama al número real de fila
df_books.append(df_books).iloc[550:555]
````
